```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the World Bank URL
url = "C:\\Users\\NIKITHA\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_306442\\API_SP.POP.TOTL_DS2_en_csv_v2_306442.csv"
df = pd.read_csv(url, skiprows=4)

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Select the relevant columns (Country Name and Population in the most recent year, e.g., 2022)
df_recent = df[['Country Name', '2022']].dropna()

# Rename columns for easier access
df_recent.columns = ['Country', 'Population']

# Sort the data by population in descending order
df_recent_sorted = df_recent.sort_values(by='Population', ascending=False)

# Display the sorted data
print(df_recent_sorted.head())

# Create a bar chart to visualize the population distribution of the top 10 most populous countries
plt.figure(figsize=(10, 6))
plt.bar(df_recent_sorted['Country'][:10], df_recent_sorted['Population'][:10], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
